```python
with open('./logs.txt', 'r') as f:
    res = []
    for line in f.readlines():
        # Epoch 0 , 
        # SSIM -0.04608755080029368 , 
        # GDL 1.18700267970562, 
        # TL 1.1409151309728622, 
        # DL 1.2456354206800462, 
        # DACC 24.90625, 
        # DGACC67.578125 
        SEP = ","
        params = line.split(SEP)
        
        prop_types = ["Epoch", "SSIM", "GDL", "TL", "DL", "DACC", "DGACC"]
        epoch_info = {}
        for prop, ptype in zip(params, prop_types):
            if len(line.strip()) > 0:
                epoch_info[ptype] = float(prop.split(ptype)[1].strip())
        res.append(epoch_info)
# print(res)
```

```python
import pandas as pd
df = pd.DataFrame(res)
df
```

```python
import matplotlib.pyplot as plt

df[["SSIM", "GDL", "TL", "DL"]].plot()
plt.show()
```
